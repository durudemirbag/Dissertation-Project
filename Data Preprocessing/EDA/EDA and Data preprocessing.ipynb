### Outlier Detection

```python
# Function to identify outliers based on the IQR rule
# Outliers = values outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
def get_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

### Correlation Matrix

#### Correlation Matrix with Smoking history

```python
# Correlation Matrix with Smoking History
# Note: One-hot encoding was used instead of label encoding to avoid misleading correlations. 
# However, this makes the results less reader-friendly and the correlations are weak, 
# so an additional matrix is plotted without smoking history.

# Temporary encoding for correlation analysis only
ohe=OneHotEncoder()
xd=ohe.fit_transform(df[["smoking_history"]]).toarray()
xd=pd.DataFrame(xd)
xd.columns=ohe.get_feature_names_out()
df_temp = df[df["gender"]!="Other"].copy() 
df_temp["gender"] = df_temp["gender"].map({"Female":0, "Male":1})
df_temp_=df_temp[['hypertension','gender','heart_disease','age', 'bmi', 'HbA1c_level','blood_glucose_level','diabetes']]
df_temp_.loc[:, xd.columns] = xd

labels = {
    'smoking_history_No Info': 'Smoking: No Info', 
    'smoking_history_current': 'Smoking: Current', 
    'smoking_history_ever': 'Smoking: Ever', 
    'smoking_history_former': 'Smoking: Former', 
    'smoking_history_never': 'Smoking: Never', 
    'smoking_history_not current': 'Smoking: Not Current',
    'gender': 'Gender',
    'HbA1c_level': 'Hemoglobin A1c Level',
    'blood_glucose_level': 'Blood Glucose',
    'heart_disease': 'Heart Disease',
    'age':'Age',
    'hypertension':'Hypertension',
    'bmi':'Body Mass Index',
    'diabetes':'Diabetes'
}
    
corr_matrix=df_temp_.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
           fmt='.2f')
plt.title('Correlation Matrix')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
